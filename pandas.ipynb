```python
import urllib.request

url = "https://www.basketball-reference.com/leagues/NBA_2015_totals.html"  # Aapka URL yahan aayega

# 1. Ek dummy Browser (User-Agent) ka Header banayein
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

try:
    # 2. Request object banayein jisme url aur headers dono hon
    request = urllib.request.Request(url, headers=headers)
    
    # 3. Ab file ko open ya download karein
    with urllib.request.urlopen(request) as response:
        data = response.read()
        print("Data downloaded successfully!")
        
except Exception as e:
    print(f"Error: {e}")
```
